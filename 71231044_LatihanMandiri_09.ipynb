```python
def perbandingan_file(file1, file2):
    with open(file1, 'r', encoding='utf-8') as f1, open(file2, 'r', encoding='utf-8') as f2:
        lines1 = f1.readlines()
        lines2 = f2.readlines()

    garis_maksimal = max(len(lines1), len(lines2))

    for i in range(garis_maksimal):
        line1 = lines1[i].strip() if i < len(lines1) else "[Tidak ada baris]"
        line2 = lines2[i].strip() if i < len(lines2) else "[Tidak ada baris]"

        if line1 != line2:
            print(f"Perbedaan pada baris {i+1}:")
            print(f"File 1: {line1}")
            print(f"File 2: {line2}")
            print("-" * 40)

file1 = "file1.txt"
file2 = "file2.txt"

perbandingan_file(file1, file2)
```

```python
def question(filename):
    questions = []
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            if "||" in line:
                soal, jawaban = line.strip().split("||")
                questions.append((soal.strip(), jawaban.strip()))
    return questions

def quiz(questions):
    print(f"Nama file: soal.txt\n")
    
    for soal, jawaban_benar in questions:
        print(soal)
        jawaban_user = input("Jawab: ").strip()
        
        if jawaban_user.lower() == jawaban_benar.lower():
            print("Jawaban benar!")
        else:
            print("Jawaban salah!")
        print("-" * 40)

file_soal = "soal.txt"
questions = question(file_soal)
quiz(questions)
```
